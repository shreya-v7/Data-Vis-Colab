```python
data_explore = df.copy()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(data_explore.State.sort_values(ascending=True), data_explore.Production)
plt.yscale('log')
plt.title('State-Wise Production: Total')
```

```python
crop = data_explore.groupby(by='Crop')['Production'].sum().reset_index().sort_values(by='Production', ascending=False).head(10)
crop 
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(crop.Crop, crop.Production)
plt.yscale('log')
plt.title('Overall Crops vs Production')
```

```python
plt.tick_params(labelsize=15)
data_explore.groupby("Year")["Production"].agg("sum").plot.bar()
```

```python
#Season vs Production
df_season=data_explore.copy()
season = df_season.groupby(by='Season')['Production'].sum().reset_index().sort_values(by='Production', ascending=False).head(10)
season
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(season.Season, season.Production,errwidth=0)
plt.yscale('log')
plt.title('Seasonal Crops vs Production')
```

```python
plt.figure(figsize=(15,8))
plt.tick_params(labelsize=10)
data_explore.groupby("Crop")["Production"].agg("count").plot.bar()
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(data=data_explore,x="State",y="Crop",hue="Season")
plt.xticks(rotation=90)
plt.show()
```

```python
df1=data_explore["Crop"].value_counts()
df1.plot(radius=3,kind="pie",autopct="%1.1f",pctdistance=0.6)
plt.tick_params(labelsize=10)
```

```python
#Step 1: Making a cross table using State Name and Crop categories
#Step 2: Visualizing cross table data in the form of stacked bar plot.
df_2 = pd.crosstab(data_explore['State'], data_explore['Crop'])
df_2
```

```python
rice_df = data_explore[data_explore["Crop"]=="Rice"]
print(rice_df.shape)
rice_df[:3]
```

```python
sns.barplot("Season","Production",data=rice_df);
```

```python
plt.figure(figsize=(13,10))
sns.barplot("State","Production",data=rice_df)
plt.xticks(rotation=90)
plt.show()
```

```python
top_rice_pro_dis = rice_df.groupby("State")["Production"].sum().reset_index().sort_values(
    by='Production',ascending=False)
top_rice_pro_dis[:10]
sum_max = top_rice_pro_dis["Production"].sum()
top_rice_pro_dis["% Production"] = top_rice_pro_dis["Production"].map(lambda x:(x/sum_max)*100)
top_rice_pro_dis[:10]
```

```python
plt.figure(figsize=(12,7))
sns.barplot("State","Production",data=top_rice_pro_dis[:10])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot("Year","Production",data=rice_df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.jointplot("Area","Production",data=rice_df,kind="reg")
```

```python
#Step 1: Find the state where Area under crop production is high and get top 10 states
#Step 2: Make a subset of the data with the above conditional data
#Step 3: Look the cultivation status of these states yearwise
df = data_explore.groupby(by='State')['Production'].sum().reset_index().sort_values(by='Production', ascending=False)
df.head()

fig, ax = plt.subplots(figsize=(25,10))
sns.barplot(df.State.head(10), df.Production.head(10), errwidth=0)
plt.title('Agricultural Area Distribution - India')
df.head(10)
```

```python
df = data_explore.groupby(by='State')['Area'].sum().reset_index().sort_values(by='Area', ascending=False)
df = df.head(5)

fig, ax = plt.subplots(figsize=(25,30), sharey='col')
count = 1

for state in df.State.unique():
    plt.subplot(len(df.State.unique()),1,count)
    sns.lineplot(data_explore[data_explore.State==state]['Year'],data_explore[data_explore.State==state]['Production'], ci=None)
    plt.subplots_adjust(hspace=0.6)
    plt.title(state)
    count+=1;
```
